```python
get_ids_query = f"""
SELECT id 
FROM morphs
WHERE morph = %s
"""

uq_w = list(dict.fromkeys([item[i] for item in b for i in range(len(item))]))

map = dict()
for w in uq_w:
    cur.execute(get_ids_query, (w,))
    id = cur.fetchone()
    map[w] = id[0]

c = tuple(tuple(sorted((map[item[0]], map[item[1]]))) for item in b)

remove_relations_query = f"""
DELETE FROM external_relations
WHERE (morph_1_id, morph_2_id) = (%s, %s)
"""

cur.executemany(remove_relations_query, c)
conn.commit()
```
