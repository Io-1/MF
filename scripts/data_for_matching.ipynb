```python
select_morphs_clusters_query = f"""
SELECT m1.morph, m2.morph
FROM external_relations er
LEFT JOIN morphs m1
  ON m1.id = er.morph_1_id
LEFT JOIN morphs m2
  ON m2.id = er.morph_2_id
"""
```

```python
cur.execute(select_morphs_clusters_query)

items = cur.fetchall()
```

```python
items
```

```python
data = {"pairs": [[pair[0], pair[1]] for pair in items]}
```

```python
import os
import json

# 1) ensure the output directory exists
os.makedirs("matching", exist_ok=True)

# 3) write it out with pretty formatting and Unicode support
with open("matching/data.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("Wrote matching/data.json")
```
